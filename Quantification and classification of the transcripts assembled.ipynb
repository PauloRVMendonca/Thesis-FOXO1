```python
# importing packages
import pandas as pd
import glob,os
  
folder_path = './Thesis'
file_list = glob.glob(folder_path + "/*.tmap")
main_dataframe = pd.DataFrame([])

for i in file_list:
    data = pd.read_csv(i,delimiter='\t')
    data['Run'] = os.path.basename(i).split('.')[0]
    df = pd.DataFrame(data)
    main_dataframe = pd.concat([main_dataframe,df],axis=0)
main_dataframe['Ensemble'] = main_dataframe['ref_id'].str[:15]
main_dataframe
```

```python
df_final=df2.groupby(['Genotype','Gene Stable ID version','Transcript stable ID']).size().reset_index(name='counts')
df_final= df_final[df_final['counts'] >= 3]
df_final
```

```python
column_names = ['Chr', 'Start', 'End','Gene Stable ID version','Gene name','Strand','Type'] 
rna_types=pd.read_csv('./Thesis/types_rna.bed',delimiter='\t',header= None,names=column_names)
rna_types
```

```python
df3= pd.merge(df_final, rna_types, on=['Gene Stable ID version'],how='inner')

# Sample DataFrame with a 'type' column

# Create the 'final' column based on the rules
df3['final'] = df3['Type'].apply(lambda x: 'small RNAs' if x.endswith('RNA') else ('pseudogene' if x.endswith('pseudogene') else x))

# Update 'final' column for specific categories
df3.loc[df3['Type'] == 'protein_coding', 'final'] = 'protein_coding'
df3.loc[df3['Type'] == 'lncRNA', 'final'] = 'lncRNA'
df3.loc[~df3['Type'].isin(['protein_coding', 'lncRNA', 'tRNA', 'snRNA', 'rRNA', 'piRNA','snoRNA','misc_RNA ','snRNA','Mt_rRNA','Mt_tRNA ' , 'processed_pseudogene', 'transcribed_processed_pseudogene','unprocessed_pseudogene','transcribe_unitary_pseudogene','rRNA_pseudogene','TR_V_pseudogene','translated_processed_pseudogene','transcribed_unprocessed_pseudogene']), 'final'] = 'other'

# Print the updated DataFrame
df3
```

```python
df4=df3.groupby(['Genotype','final']).size().reset_index(name='counts')
df5=df4.groupby(['Genotype','final']).agg({'counts': 'sum'})
df6=df5.groupby(level=0).apply(lambda x: 100*x/float(x.sum())).reset_index()
df7=df6.pivot(columns = 'final', index = ['Genotype'], values = 'counts')
df8=df7.reset_index()
df9=df8.set_index('Genotype')
df4
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# plot data in stack manner of bar type
df9.plot(kind='bar', stacked=True,colormap = plt.cm.hot, alpha = 0.8,
                           width = 0.4, edgecolor = 'black', linewidth = 0.6, clip_on = True)

plt.legend(ncol = 1, title = 'Type', title_fontsize = 7,
           loc = 0, bbox_to_anchor = (1, 1), frameon= False, fontsize = 7);
plt.title('Types of Transcripts', fontsize = 8)
plt.ylabel('Frequency (%)', fontsize = 8);
plt.xlabel('')
plt.xticks(rotation = 90, fontsize = 7, ha="center"); plt.yticks(fontsize = 8);
plt.tick_params(axis = 'y', labelsize = 7)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('Types of transcripts.png', format='png', dpi=600)
plt.show()
```
